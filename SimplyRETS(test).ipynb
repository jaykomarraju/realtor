```python
import requests
from requests.auth import HTTPBasicAuth
```

```python
# Set up the API credentials for the request
username = 'simplyrets'
password = 'simplyrets'
```

```python
# Set up the parameters for the API request
params = {
    'postalCode': '90210',  # replace with the actual zip code
}
```

```python
# Send a GET request to the SimplyRETS API
response = requests.get('https://api.simplyrets.com/properties', auth=HTTPBasicAuth(username, password), params=params)
```

```python
# Parse the JSON response
data = response.json()
```

```python
# TODO: Process the data to find the properties with the largest price cuts
```

```python
# Calculate the price cut for each property and store it in a new field
for property in data:
    if property['originalListPrice'] is not None:
        property['priceCut'] = property['originalListPrice'] - property['listPrice']
    else:
        property['priceCut'] = 0  # Assigning 0 if originalListPrice is None

# Sort the properties by the price cut in descending order
data.sort(key=lambda x: x['priceCut'], reverse=True)

# Print the properties with the largest price cuts
for property in data[:10]:  # Adjust this number to print more or fewer properties
    print(f"Property {property['listingId']} had a price cut of ${property['priceCut']}")
```
